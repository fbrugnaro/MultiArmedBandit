```python
import main
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
paths = ['Data/ydata']
main.read_news_data(paths)
```

```python
results = []
for t, event in enumerate(main.events):
    user = event[2][:5]
    max = np.argmax(user)
    v = user[max]
    results.append(v)
```

```python
plt.hist(results, weights=np.ones(len(results)) / len(results))
plt.xlabel('User Features');
plt.title('User Features Density')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.show()
```
